## 1.1 Q7 & 8 What programming languages do you use on a regular basis? (Select all that apply) and what would you recommend?

```python
df_q7 = data[['Q7_Part_1',
 'Q7_Part_2',
 'Q7_Part_3',
 'Q7_Part_4',
 'Q7_Part_5',
 'Q7_Part_6',
 'Q7_Part_7',
 'Q7_Part_8',
 'Q7_Part_9',
 'Q7_Part_10',
 'Q7_Part_11',
 'Q7_Part_12']]
df_q7.drop(index = 0,inplace = True)
```

```python
# data preprocessing
df_q7_encode = pd.get_dummies(df_q7)
```

```python
df_q7_encode.head()
```

```python
#caluclate the sum of each language used
df_q7_encode.sum()
```

```python
import matplotlib.pyplot as plt
fig,ax = plt.subplots(figsize = (10,6))

labels = ['Python','R','SQL','C','C++','Java','Javascript','Julia','Swift','Bash','MATLAB','None']
ax.set_ylabel('Number of people')
ax.set_title('Comparsion of Programming languages people use on a regular basis')
ax.set_xticklabels(labels)
bar = ax.bar(df_q7_encode.columns,height = df_q7_encode.sum())


def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')
autolabel(bar)
plt.show()
```
